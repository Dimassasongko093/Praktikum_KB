## Import Pandas Package
Import pandas sebagai pd
    
Dari matplotlib import sebagai plt
        
Import seaborn sebagai sns

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Membaca Dataset

df untuk membaca dataset 

```python
df = pd.read_csv("card_data.csv")
df
```

```python
df.dtypes
```

## Berikut 4 tipe Grafik visualiasi Data yang digunakan untuk data Yu-Gi-Oh! Trading Cards Dataset


## Grafik Scatter Plot -------------------------------------------------------------------------------------------------------
diagram grafis yang dibangun dari dua sumbu X dan Y (variabel X dan variabel
Y). Scatter plot menunjukan hubungan antara dua variabel.

```python
sns.scatterplot(data = df, x='Attribute', y = 'ATK')
plt.title("Persebaran data ukuran Attack dengan Attribute yang beda")
plt.show()
```

## Grafik Boxplot ------------------------------------------------------------------------------------------------------------

grafik ini digunakan untuk melihat persebaran data dan mencari outliers.

```python
sns.boxplot(x= df['ATK'], y=df['Attribute'])
plt.show()
```

## Grafik Pie -------------------------------------------------------------------------------------------------------------------

Grafik Pie cocok digunakan pada data
categorical. Grafik ini biasa digunakan untuk melihat distribusi.

```python
df['Attribute'].value_counts()
```

```python
labels = df['Attribute'].value_counts().rename_axis('Attribute').reset_index(name='Jumlah')
labels
```

```python
plt.figure(figsize=(8,8), dpi=300)
plt.pie(df['Attribute'].value_counts(), labels=labels['Attribute'], autopct='%.1f%%', startangle=90)
plt.show()
```

## Grafik Heatmap -----------------------------------------------------------------------------------------------------------
Grafik ini biasa digunakan untuk menunjukkan korelasi antar variable, dengan cara menampilkan
data dengan cara representasi warna yang berbeda-beda dan semakin tinggi angka suatu kelompok data maka warnanya akan
semakin gelap.

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), cmap='Oranges')
ax.set_title('Korelasi antar atribut DEF, ATK dan Level pada data Yu-Gi-Oh! Trading Cards Dataset')
plt.show()
```
